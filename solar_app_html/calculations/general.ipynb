```python
#Average prices for PV in 2025 

avg_yearly_production_dk = 175 #kWh/m2/year

small_pv_capex = 72500 #kr
small_pv_opex = 0.01 * small_pv_capex #kr/year
small_pv_kWp = 4 #kWp
small_yearly_production = 3500 #kWh/year

medium_pv_capex = 102500 #kr
medium_pv_opex = 0.01 * medium_pv_capex #kr/year
medium_pv_kWp = 7 #kWp

large_pv_capex = 142500 #kr
large_pv_opex = 0.01 * large_pv_capex #kr/year
large_pv_kWp = 10 #kWp


battery_storage_capex = 40000 #kr

#Financial parameters
r = 0.03 #    discount rate
N_pv = 30 #years (Lifetime of the PV system)
N_battery = 10 #years (Lifetime of the battery storage)
CRF_pv = r * (1 + r) ** N_pv / ((1 + r) ** N_pv - 1) # Capital Recovery Factor
CRF_battery = r * (1 + r) ** N_battery / ((1 + r) ** N_battery - 1) # Capital Recovery Factor for battery storage

#reference values for yearly savings on the electricity bill
yearly_savings_small_pv = 7000 #kr (electricity price 2 kr/kWh)
yearly_savings_medium_pv = 10500 #kr (electricity price 2 kr/kWh)
yearly_savings_large_pv = 14000 #kr (electricity price 2 kr/kWh)


#Payback period calculation
payback_period_small_pv = small_pv_capex / (yearly_savings_small_pv - small_pv_opex)
payback_period_medium_pv = medium_pv_capex / (yearly_savings_medium_pv - medium_pv_opex)
payback_period_large_pv = large_pv_capex / (yearly_savings_large_pv - large_pv_opex)
print(f"Payback period for small PV system: {payback_period_small_pv:.2f} years")
print(f"Payback period for medium PV system: {payback_period_medium_pv:.2f} years")
print(f"Payback period for large PV system: {payback_period_large_pv:.2f} years\n")




```

### Case 1.1: Small PV system without battery storage

```python
# Calculate the annual cost of the PV system
annual_cost_small_pv = small_pv_capex * CRF_pv + small_pv_opex

# Calculate the net savings
net_savings_small_pv = yearly_savings_small_pv - annual_cost_small_pv


print(f"Case 1: Small PV system (4 kWp)")
print(f"Annual cost: {annual_cost_small_pv:.2f} kr")
print(f"Net savings: {net_savings_small_pv:.2f} kr\n")
```

### #Case 1.2: Small PV system with battery storage

```python
# Calculate the total capex for the PV system with battery storage
total_capex_small_pv_battery = small_pv_capex + battery_storage_capex
opex_small_pv_battery = small_pv_opex + 0.01 * battery_storage_capex # Assuming 1% of battery cost as annual OPEX

# Calculate the annual cost of the PV system with battery storage
annual_cost_small_pv_battery = (small_pv_capex * CRF_pv + battery_storage_capex * CRF_battery + opex_small_pv_battery)

# Calculate the net savings for the PV system with battery storage
net_savings_small_pv_battery = yearly_savings_small_pv - annual_cost_small_pv_battery

print(f"Case 1.1: Small PV system with battery storage")
print(f"Annual cost: {annual_cost_small_pv_battery:.2f} kr")
print(f"Net savings: {net_savings_small_pv_battery:.2f} kr\n")
```

### Case 2: Medium PV system 

```python
# Calculate the annual cost of the PV system
annual_cost_medium_pv = medium_pv_capex * CRF_pv + medium_pv_opex


#calculate net savings for medium PV system without battery storage
net_savings_medium_pv = yearly_savings_medium_pv - annual_cost_medium_pv

print(f"Case 2: Medium PV system (7 kWp)")
print(f"Annual cost of medium PV system without battery storage: {annual_cost_medium_pv:.2f} kr/year\n")
print(f"Net savings for medium PV system without battery storage: {net_savings_medium_pv:.2f} kr/year\n")
```
